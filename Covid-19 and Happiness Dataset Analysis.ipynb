```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
#header Files
```

```python
#import Dataset Covid-19_confirmed_data.csv
corona_dataset_csv=pd.read_csv('C:\\Users\\Abhi\\Desktop\\Covid-19 Data Analysis\\covid19_Confirmed_dataset.csv')
print(corona_dataset_csv.shape)
corona_dataset_csv.head(10)
```

```python
#drop the useless columns : Lat,Long
corona_dataset_csv.drop(['Lat','Long'],axis=1,inplace=True)
print(corona_dataset_csv.shape)
```

```python
corona_dataset_csv.head(10)
```

```python
#Aggregating the rows by the country
corona_dataset_aggregated=corona_dataset_csv.groupby('Country/Region').sum()
corona_dataset_aggregated.head()
```

```python
#Visualizing data related to country
corona_dataset_aggregated.loc['India'].plot()
corona_dataset_aggregated.loc['Italy'].plot()
plt.legend()
```

```python
corona_dataset_aggregated.loc['China'][:3].plot()
```

```python
corona_dataset_aggregated.loc['India'].diff().plot()
corona_dataset_aggregated.loc['China'].diff().plot()
corona_dataset_aggregated.loc['Spain'].diff().plot()
corona_dataset_aggregated.loc['Italy'].diff().plot()
plt.legend()
```

```python
print(corona_dataset_aggregated.loc['India'].diff().max())
print(corona_dataset_aggregated.loc['China'].diff().max())
print(corona_dataset_aggregated.loc['Spain'].diff().max())
```

```python
#calculating the maximum infection rate for each country
countries=list(corona_dataset_aggregated.index)
max_infection=[]
for c in countries:
    max_infection.append(corona_dataset_aggregated.loc[c].diff().max())
max_infection
```

```python
#adding the calculated max_infection column
corona_dataset_aggregated["max_infection_rate"]=max_infection
corona_dataset_aggregated.head()
```

```python
#making new DataFrame with only important column
corona_data=pd.DataFrame(corona_dataset_aggregated['max_infection_rate'])
print(corona_data.shape)
corona_data.head()
```

```python
#importing worldwide_happiness_report.csv dataset




happiness_report_csv=pd.read_csv('C:\\Users\\Abhi\\Desktop\\Covid-19 Data Analysis\\worldwide_happiness_report.csv')
print(happiness_report_csv.shape)
happiness_report_csv.head()
```

```python
#drop the useless columns from dataset
useless_cols=['Overall rank','Score','Generosity','Perceptions of corruption']
happiness_report_csv.drop(useless_cols,axis=1,inplace=True)
print(happiness_report_csv.shape)
happiness_report_csv.head()
```

```python
happiness_report_csv.set_index('Country or region',inplace=True)
happiness_report_csv.head()
```

```python
print(happiness_report_csv.shape)
print(corona_data.shape)
```

```python
#Now Joining the both data sets for analysis

data=corona_data.join(happiness_report_csv,how='inner')
print(data.shape)
data.head()
```

```python
#perform the correlation operation 

data.corr()
```

```python
# performing the visualizing techniques to understand the dataset

x=data['GDP per capita']
y=data['max_infection_rate']
sns.scatterplot(x,y)
```

```python
# above graph is not so meaningful so make another graph with scaling
sns.scatterplot(x,np.log(y))
```

```python
sns.regplot(x,np.log(y))#here it is visible that both feature has +ve correlation
```

```python
# Now ploting the graph for every columns
x_axis=data.columns[1:]
y_axis=data['max_infection_rate']
for i in x_axis:
    x=data[i]
    sns.regplot(x,np.log(y_axis))
    plt.show()
```
